# Mở đầu

Mật mã học là một lĩnh vực liên quan đến các kỹ thuật ngôn ngữ và toán học để đảm bảo an toàn thông tin, cụ thể là trong thông tin liên lạc. Trong lịch sử, mật mã học gắn liền với quá trình mã hóa, điều này có nghĩa là nó gắn với các cách thức để chuyển đổi thông tin từ dạng này sang dạng khác, làm cho thông tin trở thành dạng không thể đọc được nếu như không có các kiến thức bí mật. Quá trình mã hóa được sử dụng chủ yếu để đảm bảo tính bí mật của các thông tin quan trọng, chẳng hạn trong công tác tình báo, quân sự hay ngoại giao cũng như các bí mật về kinh tế, thương mại. 

Trong những năm gần đây, lĩnh vực hoạt động của mật mã hóa đã được mở rộng: mật mã hóa hiện đại cung cấp cơ chế cho nhiều hoạt động hơn là chỉ duy nhất việc giữ bí mật và có một loạt các ứng dụng như: chứng thực khóa công khai, chữ ký số, bầu cử điện tử hay tiền điện tử. Ngoài ra, những người không có nhu cầu thiết yếu đặc biệt về tính bí mật cũng sử dụng các công nghệ mật mã hóa, thông thường được thiết kế và tạo lập sẵn trong các cơ sở hạ tầng của công nghệ tính toán và liên lạc viễn thông.

Một trong những ứng dụng của mật mã học và an toàn thông tin là chữ ký số. Với đặc điểm là đơn giản cho người sử dụng mà vẫn đảm bảo được tính bảo mật, kỹ thuật sử dụng chữ ký số là một trong những kỹ thuật được sử dụng phổ biến, đa dạng trong hầu hết các lĩnh vực, nhất là Tài chính, Ngân hàng, Kế toán…Vì lý do đó, chúng em quyết định nghiên cứu về “Chữ ký số RSA sử dụng hàm băm SHA-256”.


# Mô tả, mã hóa và giải mã RSA

### Mô tả

Trong mật mã học, RSA là một thuật toán mật mã hóa khóa công khai. Đây là thuật toán đầu tiên phù hợp với việc tạo ra chữ ký điện tử đồng thời với việc mã hóa. RSA đang được sử dụng phổ biến trong thương mại điện tử và được cho là đảm bảo an toàn với điều kiện độ dài khóa đủ lớn.

Thuật toán RSA có hai khóa: 
- Khóa công khai (hay khóa công cộng): được công bố rộng rãi cho mọi người và được dùng để mã hóa.
- Khóa bí mật (hay khóa cá nhân): dùng để giải mã.
    
Giả sử Alice và Bob cần trao đổi thông tin bí mật thông qua Internet. Với thuật toán RSA, Alice đầu tiên cần tạo ra cho mình cặp khóa gồm:
- Khóa công khai gồm (n,e)
- Khóa bí mật gồm (n,d)



### Mã hóa

Giả sử Bob muốn gửi đoạn thông tin M cho Alice. 
<ol>
<li>Bob chuyển M thành một số $m<n$ theo một hàm có thể đảo ngược (từ m có thể xác định lại M) được thỏa thuận trước.</li> 
<li>Bob có m và biết n cũng như e do Alice gửi. Bob sẽ tính c là bản mã hóa của m theo công thức: C = $m^e$(mod n).</li>
<li>Bob gửi c cho Alice</li>
</ol>


### Giải mã
<ol>
    <li>Alice nhận c từ Bob và biết khóa bí mật d.</li>
    <li>Alice có thể tìm được m từ c theo công thức sau: $m=c^d$ (mod n).</li>
    <li>m là thông tin nhận được.</li>


<b>Dưới đây là sơ đồ của việc tạo khóa, mã hóa và giải mã RSA</b>



# Chữ ký số RSA sử dụng hàm băm SHA-256

Chữ kí số (Digital Signature) là một chuỗi dữ liệu liên kết với một thông điệp
(message) và thực thể tạo ra thông điệp. Chữ ký số khóa công khai là mô hình sử dụng các kỹ thuật mật mã để gắn với mỗi người sử dụng một cặp khóa công khai - bí mật và qua đó có thể ký các văn bản điện tử cũng như trao đổi các thông tin mật. Khóa công khai thường được phân phối thông qua chứng thực khóa công khai. Quá trình sử dụng chữ ký số bao gồm 2 quá trình: tạo chữ ký và kiểm tra chữ ký.

Giải thuật tạo ra chữ ký số (Digital Signature generation algorithm) là một
phương pháp sinh chữ ký số.

Giải thuật kiểm tra chữ ký số (Digital Signature verification algorithm) là một
phương pháp xác minh tính xác thực của chữ ký số, có nghĩa là nó thực sự được
tạo ra bởi 1 bên chỉ định.

Một hệ chữ ký số (Figital Signature Scheme) bao gồm giải thuật tạo chữ số và giải
thuật kiểm tra chữ kỹ số.
- Quá trình tạo chữ ký số (Digital Signature signing process) bao gồm:
 - Giải thuật tạo chữ ký số.
 - Phương pháp chuyển dữ liệu thông điệp thành dạng có thể ký được.
- Quá trình kiểm tra chữ ký số (Digital signature verification process) :
 - Giải thuật kiểm tra chữ ký số.
 - Phương pháp khôi phục dữ liệu từ thông điệp.

Hàm băm (Hash Funtion) là hàm toán học chuyển đổi thông điệp (message) có độ
dài bất kỳ (hữu hạn) thành một dãy bit có độ dài cố định (tùy thuộc vào thuật toán
băm). Dãy bit này được gọi là thông điệp rút gọn (message disgest) hay giá trị băm
(hash value), đại diện cho thông điệp ban đầu.

Giải thuật Euclid mở rộng, thực hiện xét vành số nguyên modulo

```python
#Thuật toán Euclid a,b không đồng thời bằng 0, trả về coprime(a,b)
def coprime(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
#Thuật toán Euclid mở rộng: a, b không đồng thời bằng 0
#Về tư tưởng là ghép quá trình tính cặp số (x,y) vào trong vòng lặp chính của thuật toán Euclide.
def extended_gcd(aa, bb):
    lastremainder, remainder = abs(aa), abs(bb)
    x, lastx, y, lasty = 0, 1, 1, 0
    while remainder:
        lastremainder, (quotient, remainder) = remainder, divmod(lastremainder, remainder)
        x, lastx = lastx - quotient*x, x
        y, lasty = lasty - quotient*y, y
    return lastremainder, lastx * (-1 if aa < 0 else 1), lasty * (-1 if bb < 0 else 1)
```

Áp dụng thuật toán Euclid mở rộng tìm số nghịch đảo trong vành Zn

```python
def modinv(a, m):
    g, x, y = extended_gcd(a, m)
    if g != 1:
        raise Exception('Modulo nghịch đảo không tồn tại!')
    return x % m    
```

### Tạo khóa trong RSA
<ul>
    <li>Tạo hai số nguyên tố lớn ngẫu nhiên p và q</li>
    <li>Tính $n=p*q$ và $\phi(n)=(p-1)(q-1)$</li>
    <li>Chọn một số ngẫu nhiên $e$ ($0<e<\phi(n)$) sao cho ƯCLN($e,\phi(n)$)= 1</li>
    <li>Tính $d=e^{-1}$ bằng cách dùng thuật toán Euclid</li>
    <li>$n$ và $e$ làm kháo công khai</li>
    <li>$d$ làm khóa bí mật</li>
</ul>

```python
def generate_keypair(p, q):
    n = p * q
    
   #Tính giá trị hàm số Euler 'phi'
    phi = (p-1) * (q-1)

    #Chọn số nguyên e sao cho e và phi(n) là hai số nguyên tố cùng nhau
    e = randint(2,phi);

    #Sử dụng thuật toán Euclid để xác minh rằng e và phi(n) là số nguyên tố
    g = coprime(e, phi)
  
    while g != 1:
        e = randint(2,phi)
        g = coprime(e, phi)

    #Sử dụng thuật toán Euclid mở rộng để tạo khóa riêng tư (key private)
    d = modinv(e, phi)

    #Trả về giá trị khóa riêng tư và khóa công khai
    return ((e, n), (d, n))
```

### Sử dụng hàm băm SHA-256
SHA-256 tên viết tắt của Secure Hash Algorithm 256-bit là Thuật toán băm bảo mật 256 bit và dùng để tạo ra các hàm băm không thể đảo ngược và duy nhất. Số lượng hàm băm có thể có càng lớn, thì xác suất để hai giá trị sẽ tạo ra cùng một giá trị băm càng nhỏ.

<b>Quy trình mã hóa bên gửi + tạo chữ ký số</b>
<ol>
    <li>Bob tạo một tin nhắn (người dùng nhập)</li>
    <li>Bob tạo giá trị băm cho thông báo (SHA-256 với đầu vào có độ dài tùy ý).</li>
    <li>Bob sử dụng giá trị băm cùng với khóa riêng làm đầu vào cho thuật toán RSA.</li>
    <li>Bob gửi tin nhắn có đính kèm chữ ký RSA.</li>
</ol>  




```python
#Sử dụng hàm băm SHA-256   
def hashFunction(message):
    hashed = sha256(message.encode("UTF-8")).hexdigest()
    return hashed
```

```python
#Tạo và tính giá trị băm
hashed = hashFunction(message);hashed
```

```python
#Mã hóa đoạn message
def encrypt(privatek, plaintext):
    #Giải nén khóa
    key, n = privatek

    #Chuyển đổi từng chữ cái trong bản rõ thành các số dựa trên ký tự 
    #bằng cách sử dụng a^b mod m
            
    numberRepr = [ord(char) for char in plaintext]
    print("Biểu diễn số trước khi mã hóa: ", numberRepr)
    cipher = [pow(ord(char),key,n) for char in plaintext]
    
    #Trả về mảng cipher
    return cipher
```

```python
#Xuất ra đoạn tin nhắn đã được mã hóa của Bob
print("Mã hóa đoạn message bằng khóa cá nhân ", private ," . . .")
encrypted_msg = encrypt(private, hashed)   
print("Đoạn message đã được băm và mã hóa là: ")
print(''.join(map(lambda x: str(x), encrypted_msg)))
```

<b>Quy trình giải mã, xác nhận chữ ký (bên nhận)</b>
<ol>
    <li>Alice nhận được tin nhắn có chữ ký.</li>
<li>Alice tính toán giá trị băm cho tin nhắn đã nhận, giống như Bob đã làm trong bước 2 của quá trình ký (SHA-256 với đầu vào có độ dài tùy ý).</li>
    <li>Alice cung cấp mã hóa đã nhận và khóa công khai của Bob làm đầu vào để giải mã.</li>
<li>Alice so sánh giải mã RSA với giá trị băm được tính toán của cô ấy cho tin nhắn đã nhận. Nếu thuật toán trả về kết quả là chữ ký hợp lệ (trong trường hợp này là do hàm băm được giải mã của Alice khớp với hàm băm ban đầu của Bob), Alice đảm bảo rằng thư đã được ký bởi Bob, vì không ai khác có khóa riêng của anh ấy, vì vậy không ai khác có thể đã tạo một chữ ký cho thư này mà có thể được xác minh cho thư này bằng cách sử dụng khóa của Bob.</li>
</ol>   




```python
#Giải mã 
def decrypt(publick, ciphertext):
    #Giải nén bằng khóa công khai
    key, n = publick
       
    #Tạo bản rõ dựa trên bản mã và khóa bằng cách sử dụng a^b mod m
    numberRepr = [pow(char, key, n) for char in ciphertext]
    plain = [chr(pow(char, key, n)) for char in ciphertext]

    print("Biểu diễn mã được giải dạng số: ", numberRepr)
    
    #Trả về mảng dưới dạng chuỗi 
    return ''.join(plain)
```

```python
#Xác minh lại chữ ký và kết luận   
def verify(receivedHashed, message):
    ourHashed = hashFunction(message)
    if receivedHashed == ourHashed:
        print("Xác minh thành công! ", )
        print(receivedHashed, " = ", ourHashed)
    else:
        
        print("Xác minh không thành công")
        print(receivedHashed, " != ", ourHashed)
        
```
